# Taylor Series 

This lecture was mostly about the Taylor Series. The Taylor Series gives a nice way to approximate non-linear functions using polynomial expansion around some point $c$ which we may consider the center around which we want an approximation of the function. This is useful in computing some non-linear functions in computers like the sin, or the cos functions, which can be extremely helpful in the case we have small angles.

The Taylor Series is defined as:
$$
f(x) = \sum_{n=0}^{\infty}\frac{f^{(n)}(c)(x-c)^{n}}{n!}
$$

The **Taylor Theorem** is defined as:
$$
f(x) = \sum_{n=0}^{m}\frac{f^{(n)}(c)(x-c)^{n}}{n!} + \frac{f^{(m+1)}(\xi)(x-\xi)}{(m+1)!}
$$

In order to simulate the Taylor Series, we must first import the necessary libraries.

```python
import numpy as np
from math import factorial
import matplotlib.pyplot as plt
```

Prepare the $x$ values, and set up the figure to plot

```python
x = np.linspace(-np.pi, np.pi, 200)
plt.figure(figsize=(10, 8))
```

We start at $n = 0$, the first-order approximation:

```python
n = 0
```

In order to approximate the taylor series expansion for $\sin(x)$, we must first declare the $y$ values.
Then, the summation can be done - it must be recalled that for any even $n$-th term, the $\sin(x)$ will be 0.

The taylor series expansion of $\sin(x)$ can be given as follows:
$$
    f(x) = \sum_{n=0}^{m} \frac{(-1)^{n} \cdot x^{2n+1}}{2n+1}
$$

```python
n = n / 2

y = np.zeros(len(x))
for i in np.arange(0, int(n) + 1):
    y = y + (-1)**i * (x) ** (2*i+1) / factorial(2*i+1)

plt.plot(x, y, label = 'First Order')
```

This is about what we expected for a first-order approximation. This is a polynomial of only one order, so it of course will only be a coefficient slope. We may then proceed with further approximations by adding further polynomial terms.

```python
n = 2
n = n / 2

y = np.zeros(len(x))
for i in np.arange(0, int(n) + 1):
    y = y + (-1)**i * (x) ** (2*i+1) / factorial(2*i+1)

plt.plot(x, y, label = 'Second Order')
```

We amp it up to the $11$-th order.

```python
n = 10
n = n / 2

y = np.zeros(len(x))
for i in np.arange(0, int(n) + 1):
    y = y + (-1)**i * (x) ** (2*i+1) / factorial(2*i+1)

plt.plot(x, y, label = 'Second Order')
```

And at the $10$-th order.. the approximation is pretty close!
